## View Results

After training completes, check the outputs:

```python
import os
from pathlib import Path

# Check output directories
output_dirs = [
    "outputs/models",
    "outputs/plots",
    "outputs/reports"
]

for dir_path in output_dirs:
    path = Path(dir_path)
    if path.exists():
        files = list(path.iterdir())
        print(f"\n{dir_path} ({len(files)} files):")
        for f in files[:5]:  # Show first 5 files
            print(f"  - {f.name}")
    else:
        print(f"\n{dir_path}: Directory not found")
```

## Display Training Plots

```python
from IPython.display import Image, display
from pathlib import Path

model_name = os.environ.get('MODEL_NAME', 'resnet')

# Display training history plot
history_plot = Path(f"outputs/plots/{model_name}_history.png")
if history_plot.exists():
    print(f"\nTraining History - {model_name.upper()}")
    display(Image(filename=str(history_plot)))

# Display confusion matrix
cm_plot = Path(f"outputs/plots/{model_name}_confusion_matrix.png")
if cm_plot.exists():
    print(f"\nConfusion Matrix - {model_name.upper()}")
    display(Image(filename=str(cm_plot)))
```
